```python
from pyspark.sql import SparkSession

spark_session = SparkSession.builder.appName("spark_practice").getOrCreate()
```

```python
df = spark_session.read.csv(header=True, inferSchema=True, path="sample_data_pandas.csv")
```

```python
df.show()
```

```python
df.printSchema()
```

```python
df.drop("Name").show()
```

```python
df.na.drop().show()
```

```python
## how = any

df.na.drop(how="any").show()
```

```python
df.na.drop(how="all").show()
```

```python
df.na.drop(how="any", thresh=4).show() ### threshold value set for atlleast mentioned non null value should be there in row
```

```python
df.na.drop(how="any", subset=["Age"]).show()
```

```python
df.na.fill(value =1000000).show()
```

```python
df.na.fill(value =1000000,subset=["Age"]).show()
```

```python
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=["Age", "Salary","Experience"],
    outputCols=["{}_imputed".format(c) for c in ["Age", "Salary","Experience"]]
).setStrategy("mean")
```

```python
imputer.fit(df).transform(df).show()
```

```python
from pyspark.ml.feature import Imputer

imputer = Imputer(
    inputCols=["Age", "Salary","Experience"],
    outputCols=["{}_imputed".format(c) for c in ["Age", "Salary","Experience"]]
).setStrategy("median")
```

```python
imputer.fit(df).transform(df).show()
```
